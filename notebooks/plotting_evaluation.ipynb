```python
with open("../outputs/AudioMNIST_gender_evaluation_fft_1000.pkl", "rb") as f:
    evaluation_gender = pickle.load(f)
with open("../outputs/AudioMNIST_digit_evaluation_fft_1000.pkl", "rb") as f:
    evaluation_digit = pickle.load(f)
```

```python
evaluation = evaluation_digit

model_name = "freqrise_200_10000_dropprob_0.5"

x_axis = evaluation["deletion curves"][model_name]["quantiles"]
y_axis = evaluation["deletion curves"][model_name]["mean_prob"]
print("Complexity score: ", evaluation["complexity scores"][model_name])
print("Faithfulness: ", evaluation["deletion curves"][model_name]["AUC"])
plt.plot(x_axis, y_axis, marker='o')
plt.xlabel("Quantiles")
plt.ylabel("Mean Probability")
plt.title(f"Deletion Curve for {model_name}")
plt.grid()
plt.show()
```
