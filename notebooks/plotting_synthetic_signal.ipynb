```python
import pickle
import matplotlib.pyplot as plt
import torch
from torch.fft import rfft
```

```python
with open(f'samples/synthetic_test_dataset.pkl', 'rb') as f:
    samples = pickle.load(f)

signals = samples['signals']
labels = samples['labels']

for i in range(len(signals)):
    signal = signals[i]
    label = labels[i]
    print(f"Label {i}: {label}")
    # Create a time vector
    N = 400
    fs = 400
    t = torch.arange(0, N) / fs
    # Compute the real FFT
    fft_result = rfft(torch.Tensor(signal))
    # Compute frequency axis
    frequencies = torch.fft.rfftfreq(N, d=1/fs)
    # Plot the magnitude spectrum and time signal side-by-side
    fig, axs = plt.subplots(1, 2, figsize=(16, 4))

    # Time signal
    axs[0].plot(t.numpy(), signal)
    axs[0].set_xlabel('Time (s)')
    axs[0].set_ylabel('Magnitude')
    axs[0].set_title('Test Signal')
    axs[0].grid()

    # Magnitude spectrum
    axs[1].plot(frequencies.numpy(), torch.abs(fft_result).numpy())
    axs[1].set_xlabel('Frequency (Hz)')
    axs[1].set_ylabel('Magnitude')
    axs[1].set_title('FFT of Test Signal')
    axs[1].grid()


    plt.tight_layout()
    plt.show()
```
